```python
import pandas as pd
import Levenshtein
import os
```

```python
# Chemin du dossier contenant les fichiers CSV
dossier = '/Users/ozer/miniconda3/envs/proj1/ift3700/Données'
fichier_final = '/Users/ozer/miniconda3/envs/proj1/ift3700/mon_dataframe1.csv'
df2 = pd.read_csv(fichier_final)
    # Parcourir tous les fichiers dans le dossier
i = 1
for fichier in os.listdir(dossier):
    if fichier.endswith('.csv'):
        chemin_complet = os.path.join(dossier, fichier)
        
        # Lire le fichier CSV
        df1 = pd.read_csv(chemin_complet)

        for index in range(len(df1)):
            # Utiliser .iloc pour accéder à la valeur par position
            cell_pays = df1.iloc[index, 0]
            cell_valeur = df1.iloc[index, 1]
            temp_similarite = 0
            temp_index = 0
            
            for index2 in range(len(df2)):
                cell_pays2 = df2.iloc[index2, 0]
                
                # Comparer les valeurs des cellules
                if Levenshtein.ratio(cell_pays, cell_pays2) > 0.95:
                    df2.iloc[index2, i] = cell_valeur
                    print(cell_pays, cell_pays2)
                    break
                if Levenshtein.ratio(cell_pays, cell_pays2) > 0.80 and Levenshtein.ratio(cell_pays, cell_pays2) > temp_similarite:
                    # Remplacer la valeur de la cellule
                    print(cell_pays, cell_pays2)
                    temp_index = index2
                    temp_similarite = Levenshtein.ratio(cell_pays, cell_pays2)
                    temp_cell_pays = cell_pays
                    
            if temp_similarite> 0.80:
                print(cell_pays,temp_cell_pays)
                df2.iloc[temp_index, i] = cell_valeur
    
    i = i + 1    
            # Enregistrer les modifications
    df2.to_csv(fichier_final, index=False)
```
